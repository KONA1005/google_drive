# 데이터 다뤄보기
- 행 열 핸들링 해보자

```python
 juice['Sold'] = 0
 print(juice.head(3))
```

```python
 juice['Sold'] =  juice['Lemon'] +  juice['Orange']
 print(juice.head(3))
```

- 매출액 = 가격 * 판매량
- Revenue 로 구해봐라

```python
 juice['Revenue'] =  juice['Price'] *  juice['Sold']
 print(juice.head(3))
```

- 행과 열 제거 하는 함수
- drop(axis=0 or 1)
  + axis = 0 으로 설정시 행방향(index)으로 drop() 실행
  + axis = 1 로 설정시 열방향으로 drop() 실행

```python
 juice_colum_drop = juice.drop('Sold', axis = 1)
 print(juice_colum_drop.head(3))
```

```python
 juice_row_drop = juice.drop(0, axis = 0)
 print(juice_row_drop.head(3))
```

# Groupby() 정말 많이 씀
- 피벗테이블
- 목적 : 데이터 요약하려고. R도 마찬가지
- R dplyr groupby() %>% summarize()

```python
 juice.groupby(by = 'Location').count()
```

```python
import numpy as np

juice.groupby(['Location'])[['Revenue','Lemon']].agg([max, min, sum, np.mean])
# 각 로케이션에 따른 직계함수가 만들어짐
```
