이상치 다룰때 일반적인 통계 공식. 절대적 X
- IQR 쓴다. R의 boxplot 사분위값
- Q0(0, Q1(25%), Q2(50%), Q3(75%), Q4(100%)
- 이상치의 하한 경계값 : Q1 - ( 1.5 * (Q3 - Q1))
- 이상치의 상한 경계값 : Q3 + ( 1.5 * (Q3 - Q1)
- 도메인(각 비즈니스 영역, 미래 일자리)에서 바라보는 이상치의 기준(관습) 이 우선임. 통계공식은 통계공식일 뿐 

```python
sales[['Unit price']].describe()
```

```python
Q1 = sales['Unit price'].quantile(0.25)
Q3 = sales['Unit price'].quantile(0.75)

outliers_q1 = Q1 - 1.5 * (Q3-Q1)
outliers_q1
# 값안에 결측치가 없는 데이터라.. Q1보다 낮은 값을 이상치로 간주
outliers_q1 = (sales['Unit price'] < Q1)

# Q3보다 높은 값을 이상치로 간주
outliers_q3 = (sales['Unit price'] > Q3)
```

```python
# 이거 고쳐야됨 print(sales['Unit price][~(outliers_q1 | outliers_q3)])
```
